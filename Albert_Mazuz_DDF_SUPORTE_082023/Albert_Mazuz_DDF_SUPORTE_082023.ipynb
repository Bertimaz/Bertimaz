#Item 2 - Sobre Processos Internos
Imagine que a Dadosfera está passando por um upgrade e incorporando uma nova
plataforma de gerenciamento de diretório em nuvem, com SSO e recursos de ciclo de vida
do usuário. Como você organiza e implementa esses marcos para garantir uma transição

##Item 2 - Resposta
Implementar uma nova plataforma de gerenciamento de diretório em nuvem com Single Sign-On (SSO) e recursos de ciclo de vida do usuário é uma tarefa complexa. Para tanto é necessaria uma abordagem bem planejada e comunicação clara com os clientes. Abaixo está um plano de como isso poderia ser organizado:

1. Desenvolvimento de Documentação:

 Desenvolva documentação clara e abrangente, incluindo guias passo a passo, tutoriais em vídeo e FAQs sobre como usar a nova plataforma, configurar SSO e aproveitar os recursos de ciclo de vida do usuário.

2. Treinamento dos Clientes:

Ofereça sessões de treinamento ao vivo ou gravadas para os clientes aprenderem a navegar e aproveitar a nova plataforma. Isso ajuda a reduzir a curva de aprendizado e a garantir que os clientes estejam confortáveis com as mudanças.

3. Comunicação Prévia:

Comunique-se com seus clientes com antecedência sobre a atualização iminente. Isso deve ser feito da seguinte maneira:

*   Dois Meses para o Lançamento: Email para todos os clientes com documentação e links de treinamento
*   Um mês para Lançamento: Email de reforço para cliente que não iniciaram treinamento
*   Duas semanas para o Lançamento: webinars informativos.
*   Uma Semana para lançamento: notificações no painel de controle.  

4. Testes Beta:
Lançamento da versão beta para grupo seleto de clientes e identificação de problemas antes do lançamento completo. Avaliar a necessidade de lançamento gradual da nova plataforma.

5. Lançamento Gradual:
Lançamento parcial de acordo com dimensionamento da equipe de suporte de acordo com avaliação na etapa anterior.

6. Suporte Contínuo:
Mantenha uma equipe de suporte dedicada disponível para ajudar os clientes durante a transição. Isso pode ser feito através de chat ao vivo, suporte por e-mail ou uma linha direta de suporte.

7. Feedback Pós-Lançamento:
Após o lançamento, peça feedback dos clientes sobre a nova plataforma. Isso não só demonstra que você valoriza a opinião deles, mas também ajuda a identificar áreas que podem precisar de ajustes adicionais.

8. Atualização da Documentação:
À medida que os clientes começarem a usar a nova plataforma, atualize a documentação com base em perguntas frequentes e feedback para torná-la ainda mais útil e informativa.

Em resumo, uma implementação suave e eficiente da nova plataforma de gerenciamento de diretório em nuvem requer planejamento detalhado, comunicação clara e suporte dedicado. Mantendo os clientes informados, treinados e apoiados, você estará melhor preparado para garantir uma transição bem-sucedida e minimizar qualquer interrupção nas operações das empresas.

#Item 4 - Sobre SQL
Na Dadosfera, temos uma base de dados com informações fictícias de uma empresa. Foi demandado pela sua liderança de CS, que você crie análises descritivas da própria área com base nesses dados.


## Item 4 - Resposta

A visualização de Nº de funcionários por cargo foi criada a partir do seguinte comando SQL:


```
SELECT *
FROM (SELECT
        CASE
        WHEN jobrole='Human Resources' or jobrole='Research Director' or jobrole='Manager' or jobrole='Sales Representative' THEN 'Other'
        ELSE jobrole
        END as cargo,
        COUNT(jobrole) AS "Quantidade de Funcionários"
FROM TB__Z96BQ3__EMPLOYEE
group by cargo


) AS "source"
LIMIT 1048575
```
E configurado como gráfico de barras horizontais ficou da seguinte maneira:

De maneira similar foi criado a visualização de Número de funcionários pode departamento:


```
SELECT "source"."DEPARTMENT" AS "DEPARTMENT", "source"."SUM(EMPLOYEECOUNT)" AS "SUM(EMPLOYEECOUNT)"
FROM (SELECT
department,
SUM(employeecount)
FROM TB__Z96BQ3__EMPLOYEE
GROUP BY department


) AS "source"
LIMIT 1048575
```
